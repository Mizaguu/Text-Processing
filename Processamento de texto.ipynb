```python
# Extraindo textos de arquivos txt:
def extrai_txt(nome_arquivo):
    arquivo = open(nome_arquivo, 'r', encoding='utf-8')
    return arquivo.read()
```

```python
# Extraindo textos de arquivos word:
# pip install python-docx

import docx

def extrai_word(nome_arquivo):
    doc = docx.Document(nome_arquivo)
    texto = []
    for paragrafo in doc.paragraphs:
        texto.append(paragrafo.text)
    return '\n'.join(texto)

nome_arquivo = 'word_teste.docx'
print('Documento completo: \n', extrai_word(nome_arquivo))
```

```python
documento = docx.Document(nome_arquivo)
print('Número de parágrafos:', len(documento.paragraphs))
print('Primeiro parágrafo:', documento.paragraphs[1].text)

# Documentação: https://python-docx.readthedocs.io/en/latest/
```

```python
# Extraindo textos de arquivos PDF:
# pip install pypdf2

from PyPDF2 import PdfFileReader

def extrai_pdf(nome_arquivo, senha=''):
    arquivo = open(nome_arquivo, 'rb')
    leitor_pdf = PdfFileReader(arquivo)
    if senha != '':
        leitor_pdf.decrypt(senha)
    texto = []
    for i in range(0, leitor_pdf.getNumPages()):
        texto.append(leitor_pdf.getPage(i).extractText())
    return '\n'.join(texto)

nome_arquivo = 'pdf_teste.pdf'

print('Texto PDF: \n', extrai_pdf(nome_arquivo))

# Documentação: https://pythonhosted.org/PyPDF2/
```

```python
# Comparando vocabulários e textos por meio de operadores lógicos
texto1 = "Os problemas de classificação são aqueles onde se busca encontrar uma classe, dentro das possibilidades existentes."

texto2 = "Problemas de classificação são muito comuns. A busca por uma classe específica é o principal objetivo."

# Colocando todas as letras como minúsculas:
texto1 = texto1.lower()
texto2 = texto2.lower()

# Substituindo caracteres irrelevantes:
texto1.replace(',', '').replace('  ', ' ').replace('.', '')
texto2.replace(',', '').replace('  ', ' ').replace('.', '')

# Criando listas de palavras:
palavras1 = texto1.split(' ')
palavras2 = texto2.split(' ')

# Mantendo apenas palavras únicas:
palavras_unicas1 = set(palavras1)
palavras_unicas2 = set(palavras2)

# Selecionando palavras em comum:
vocabulario_comum = palavras_unicas1 & palavras_unicas2
```

```python
vocabulario_comum
```

```python
# Selecionando palavras que aparecem somente em um dos textos:
vocabulario_exclusivo = palavras_unicas1 ^ palavras_unicas2
```

```python
vocabulario_exclusivo
```

```python
# Selecionando palavras que aparecem em pelo menos um dos textos:
vocabulario_total = palavras_unicas1 | palavras_unicas2
vocabulario_total
```

```python
# Expressões Regulares
import re # pacote que lida com expressões regulares (trabalha com símbolos e caracteres especiais)
```

```python
# Utilizando a função search()
if re.search("artificial", "inteligência artificial"): 
    print("Correspondência encontrada!")
else:
    print("Não encontrada!")
```

```python
# Utilizando a função finditer
texto = "Não existe doce mais doce que o doce da batata doce"
for localizador in re.finditer("doce", texto):
    inicio = localizador.start()
    fim = localizador.end()
    print("Encontrei a palavra", texto[inicio:fim], "começando na posição", inicio, "teminando na posição", fim)
```

```python
# Usando a função search() com operadores coringa:

# O operador * significa zero ou mais vezes. 
# Quando colocado ao lado de uma letra, ele significa que aquela letra pode aparecer zero ou mais vezes.

if re.search("tel*", "intligência artificial"):
    print("Correspondência encontrada!")
else:
    print("Não encontrada!")
```

```python
# O operador + significa uma ou mais vezes. 

if re.search("tel+", "inteigência artificial"):
    print("Correspondência encontrada!")
else:
    print("Não encontrada!")
```

```python
# Trabalhando com mais de um operador
# O operador . significa qualquer coisa exceto nova linha
if re.search("te.+l", "o dente é dee"): 
    print("Correspondência encontrada!")
else:
    print("Não encontrada!")
```

```python
# O operador $ significa terminando com essa letra
if re.search("d.*e$", "o dente é dela"): 
    print("Correspondência encontrada!")
else:
    print("Não encontrada!")
```

```python
# Especificando conjuntos de caracteres:
# Conjuntos de caracteres são representados dentro de chaves []
# O circunflexo no início de chaves [^] significa qualquer coisa exceto. 
# a-z significa qualquer caractere entre as letras a e z
# 0-9 significa qualquer número entre 0 e 9
# \s significa qualquer espaço
# Portando, [^a-z0-9\s] significa qualquer coisa que não seja uma letra de a-z, um número de 0-9 ou um espaço.

if re.search("[^a-z0-9\s]", "abcdefghijklmnop%qrstuvxyz 0123456789"): 
    print("Correspondência encontrada!")
else:
    print("Não encontrada!")
```

```python
# Usando a função findall e outros caracteres para encontrar datas:
# No código abaixo, as expressões dentro dos parênteses são isoladas, ou seja, representam uma busca independente agrupada
# \d significa qualquer número
# {1,2} significa uma ou duas ocorrências. Portanto, a expressão (\d{1,2}) significa qualquer conjunto de 1 ou 2 números
# A expressão (\d{2,4}) significa qualquer conjunto de 2 a 4 dígitos 
texto = "A declaração foi feita em 16/07/2014, durante a cerimônia"
formato_data = '(\d{1,2})/(\d{1,2})/(\d{2,4})'
```

```python
if re.findall(formato_data, texto):
    print('Data encontrada:', re.findall(formato_data, texto))
else:
    print('Não encontrada!')
```

```python
# Utilizando a função sub():
# Essa função é usada para substituir uma coisa por outra. Eu preciso informar 3 parâmetros: o que quero que seja substituído,
# o que quero colocar no lugar como substituição e qual o texto/variável que irei analisar para fazer essa substituição.

dados = "O ed. estava com lotação de 15%"
print(re.sub("ed.", "edifício", dados))
```

```python
dados = "O ed. estava com lotação de 15%"
print(re.sub('[^a-z0-9\s]','',dados))
```

```python
# \w significa qualquer caractere de palavra (letras ou números)
dados = "O ed. estava com lotação de 15%"
print(re.sub('[^\w\s]','',dados.lower()))
```
